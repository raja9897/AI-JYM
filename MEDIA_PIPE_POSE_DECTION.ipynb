CALCULATE ANGLES

```python
def calculate_angle(a,b,c):
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)
    
    radians = np.arctan2(c[1]-b[1],c[0]-b[0])-np.arctan2(a[1]-b[1],a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)
    
    if angle>180.0:
        angle = 360-angle
    return angle
```

```python
# Setup
import cv2
import numpy as np
import cvzone
from cvzone.PoseModule import PoseDetector
import math

counter = 0
direction = 0

cap = cv2.VideoCapture(0)  # Use webcam feed (change 0 to another number if you have multiple cameras)
pd = PoseDetector(trackCon=0.70, detectionCon=0.70)

# Function to calculate angles and count sit-ups
def angles(lmlist, p1, p2, p3, p4, p5, p6, drawpoints):
    global counter
    global direction

    if len(lmlist) != 0:
        point1 = lmlist[p1][:2]  
        point2 = lmlist[p2][:2]
        point3 = lmlist[p3][:2]
        point4 = lmlist[p4][:2]
        point5 = lmlist[p5][:2]
        point6 = lmlist[p6][:2]

        x1, y1 = point1
        x2, y2 = point2
        x3, y3 = point3
        x4, y4 = point4
        x5, y5 = point5
        x6, y6 = point6

        if drawpoints == True:
            cv2.circle(img, (x1, y1), 10, (255, 0, 255), 5)
            cv2.circle(img, (x1, y1), 15, (0, 255, 0), 5)
            # More circles and lines...

        lefthipangle = math.degrees(math.atan2(y3 - y2, x3 - x2) - math.atan2(y1 - y2, x1 - x2))
        righthipangle = math.degrees(math.atan2(y6 - y5, x6 - x5) - math.atan2(y4 - y5, x4 - x5))

        leftHipAngle = int(np.interp(lefthipangle, [-30, 180], [100, 0]))
        rightHipAngle = int(np.interp(righthipangle, [34, 173], [100, 0]))

        left, right = leftHipAngle, rightHipAngle

        if left >= 70 and right >= 70:
            if direction == 0:
                counter += 0.5
                direction = 1
        if left <= 70 and right <= 70:
            if direction == 1:
                counter += 0.5
                direction = 0

        cv2.rectangle(img, (0, 0), (120, 120), (255, 0, 0), -1)
        cv2.putText(img, str(int(counter)), (20, 70), cv2.FONT_HERSHEY_SCRIPT_SIMPLEX, 1.6, (0, 0, 255), 7)
        # More rectangles and text...

# Main loop
while True:
    ret, img = cap.read()
    img = cv2.resize(img, (1000, 500))
    cvzone.putTextRect(img, 'AI Sit-Up Counter', [345, 30], thickness=2, border=2, scale=2.5)
    pd.findPose(img, draw=0)
    lmlist, bbox = pd.findPosition(img, draw=0, bboxWithHands=0)
    angles(lmlist, 11, 13, 15, 12, 14, 16, drawpoints=1)
    cv2.imshow('frame', img)
    if cv2.waitKey(1) == ord('q'):
        break

# Release the webcam and close all windows
cap.release()
cv2.destroyAllWindows()
```
